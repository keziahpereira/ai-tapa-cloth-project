# AI Experiments and Interpretation

This notebook explores how generative AI systems interpret digitised images from Alexander Shaw’s tapa cloth catalogue. By comparing AI-generated outputs with photographs of the original archive, the project investigates whether AI can meaningfully engage with culturally significant material objects or whether it mainly responds to visual patterns and surfaces.

I will be focusing on image interpretation, generation, and comparison in order to examine questions of authenticity, materiality, and representation within digital cultural heritage.

# AI Tools Used

The AI experiments in this project used generative AI and image interpretation tools including ChatGPT and Adobe Firefly. These tools were used to analyse digitised photographs of the tapa cloth samples and to generate new AI-produced interpretations inspired by the archive.

ChatGPT was used for image interpretation and text-based analysis of the tapa cloth photographs, while Adobe Firefly was used to generate synthetic textile images based on prompts related to barkcloth, texture, and archival material.

These tools were used as part of a collaborative creative process rather than as fully autonomous systems. The experiments focused on how AI responds to digitised cultural heritage objects and whether material and cultural qualities become simplified during image generation and interpretation.

The first experiment explored how AI interprets digitised images of tapa cloth when prompted to describe the archive visually. One of the close-up photographs taken during digitisation was uploaded into ChatGPT in order to observe how the system recognised texture, material, and pattern within the image.

One detail that stood out to me was that the AI did not meaningfully describe the bold striped and diagonal patterns present in the largest tapa sample, despite these being one of the most visually dominant features in the image. This made the response feel visually accurate at a surface level, but lacking in close material observation and specificity.

# Experiment 2: AI Image Generation

The second experiment explored how generative AI recreates and transforms digitised archival imagery. One of the photographs from the tapa cloth catalogue was uploaded into an AI image generation tool in order to produce a synthetic interpretation inspired by the original sample.

The following prompt was used:

> “Generate a digital textile inspired by this tapa cloth sample while preserving its handmade geometric patterns and archival qualities.”

The aim of this experiment was to compare the generated image with the original archival photograph and observe how the AI interpreted texture, materiality, pattern, and signs of ageing within the object.

![Comparison between original tapa cloth sample and AI-generated interpretation](images/aigenfig2.png)

*Figure 2. Comparison between the original tapa cloth sample from Alexander Shaw’s catalogue (left) and the AI-generated interpretation inspired by the archival image (right).*

## Reflection

What surprised me most about this experiment was how closely the AI-generated image resembled the original tapa cloth sample. I did not expect ChatGPT to produce something that looked almost like a direct replica, especially since the prompt only asked for a digital textile inspired by the archival image rather than an exact copy.

Although the generated image appears very similar at first glance, there are still noticeable differences when looking more closely. Certain areas of the cloth appear darker, cleaner, and more symmetrical than the original sample, and some of the irregular textures and signs of ageing visible in the archive have been simplified. Despite this, the AI was still able to recreate many of the geometric patterns and visual qualities of the tapa cloth with surprising accuracy.

This experiment raised interesting questions about digital accessibility and AI’s growing ability to reproduce archival imagery almost exactly. While this could make cultural heritage more accessible digitally, it also raises concerns about originality, authenticity, and how easily AI can replicate historical material without fully understanding its physical or cultural context.

At first glance, the generated image appears successful because it captures the warm earthy colour palette and geometric arrangement of the original textile. The repeated striped blocks and textured square shapes clearly show that the AI recognised the main visual elements present in the tapa cloth sample.  

However, when looking more closely, the differences between the original material and the AI interpretation become much more noticeable. While the AI reproduced the colours and general composition quite well, it failed to fully capture the actual texture and physical quality of the material. The original tapa cloth contains visible irregularities, layered fibres, fading, and uneven handmade marks that give it a tactile and archival feel. In comparison, the generated textile looks flatter and more digitally processed. The surface appears cleaner and more repetitive, lacking the depth and organic imperfections that make the original sample feel handmade and aged.  

This experiment made me realise that AI can imitate the visual appearance of historical textiles quite effectively from a distance, but it still struggles to recreate the material authenticity and subtle textures of handmade archival objects. Instead of producing an exact recreation, the outcome feels more like a modern digital interpretation inspired by the tapa cloth rather than the material itself.

## Conclusion

This project helped me explore how AI can interpret and recreate textile designs through digital image generation. While the AI was able to replicate colours, patterns, and the overall composition quite well, the experiments showed that it still struggled to fully capture the texture, depth, and handmade qualities of the original textiles. 

Through these experiments, I realised that AI can be a useful creative tool for inspiration and fast visual production, but it cannot completely replace the human understanding, cultural meaning, and material awareness involved in textile design. Overall, the project gave me a better understanding of both the possibilities and limitations of AI within creative practice.
