# AI Experiments and Interpretation

This notebook explores how generative AI systems interpret digitised images from Alexander Shaw’s tapa cloth catalogue. By comparing AI-generated outputs with photographs of the original archive, the project investigates whether AI can meaningfully engage with culturally significant material objects or whether it mainly responds to visual patterns and surfaces.

I will be focusing on image interpretation, generation, and comparison in order to examine questions of authenticity, materiality, and representation within digital cultural heritage.

# AI Tools Used

The AI experiments in this project used generative AI and image interpretation tools including ChatGPT and Adobe Firefly. These tools were used to analyse digitised photographs of the tapa cloth samples and to generate new AI-produced interpretations inspired by the archive.

ChatGPT was used for image interpretation and text-based analysis of the tapa cloth photographs, while Adobe Firefly was used to generate synthetic textile images based on prompts related to barkcloth, texture, and archival material.

These tools were used as part of a collaborative creative process rather than as fully autonomous systems. The experiments focused on how AI responds to digitised cultural heritage objects and whether material and cultural qualities become simplified during image generation and interpretation.
