# Digitisation and Data Collection



This notebook documents the process of digitising selected pages and tapa cloth samples from Alexander Shaw’s catalogue. The aim of this stage was to transform the physical object into digital material that could later be explored through generative AI tools and image analysis. Because the catalogue is a highly material object, the process of digitisation also raised questions about texture, preservation, detail, and what may be lost when physical cultural heritage objects are viewed digitally.

### Digitisation Process

The images used throughout this project were captured during my visit to the University of Glasgow Special Collections. While photographing the catalogue, I became much more aware of how delicate the object was physically. Because of the age and fragility of the pages, the book had to be handled very carefully and pages could only be turned in a certain way to avoid damaging the inserted tapa samples. This experience made the object feel very different from viewing a normal digitised archive online. The material qualities of the catalogue, including the texture of the barkcloth, the uneven surfaces, and the visible signs of age, became much more noticeable in person. Photographing the object also highlighted how difficult it is to fully capture these physical details digitally. Although the images recorded the visual appearance of the tapa samples, aspects such as texture, thickness, and tactility were much harder to preserve through photographs alone. This became an important part of the project because the AI tools used later only interact with the digital images rather than the physical object itself.

# Selecting Images for Digitisation

Not every page within the catalogue was photographed for this project. I focused on 3 images that showed different visual and material qualities of the tapa cloth samples. Some pages were chosen because of their detailed patterns and geometric designs, while others highlighted visible fibres, fading, layered textures, and signs of ageing within the archive.

The images were also selected based on how clearly they could communicate the physical qualities of the object once digitised. Since the later AI experiments rely entirely on digital image data, I wanted to include photographs that showed a range of surfaces, colours, and material details that might influence how generative AI interprets the cloth.

During this process, I became more aware of how the act of selecting and digitising images changes the way the archive is represented digitally. Certain details become more visible through photography, while others such as texture, thickness, and tactility remain difficult to fully capture.
