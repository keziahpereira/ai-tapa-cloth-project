# Threads of Authenticity: Human–AI Interpretation of Alexander Shaw’s Tapa Cloth Catalogue


# Introduction


For this project, I selected *Sp Coll Hunterian K.5.22, A catalogue of the different specimens of cloth collected in the three voyages of Captain Cook, to the southern hemisphere* published in London in 1787. The catalogue contains real samples of tapa cloth collected during Captain Cook’s voyages across the Pacific and later brought back to Britain.

Tapa cloth, also known as barkcloth, is a traditional textile made in several Pacific cultures from the inner bark of trees. What makes this catalogue especially unusual is that the cloth samples were inserted into the book by hand, meaning that each surviving copy is materially unique rather than identical like most printed books.

The object is held within the University of Glasgow Special Collections and exists at the intersection of colonial collecting, material culture, archival preservation, and tactile history. More than just a printed catalogue, it functions as both a historical document and a physical textile archive.

### What is Tapa Cloth?

Before starting this project, I had never really thought about how much history and meaning could exist within a piece of cloth. Tapa cloth, also known as barkcloth, is a traditional textile made across several Pacific cultures including Tonga, Samoa, Fiji, and Hawai‘i. It is created from the inner bark of trees, usually the paper mulberry tree, which is soaked, beaten by hand, and layered into sheets. Unlike woven fabric, tapa cloth is handmade through a long physical process that involves a lot of labour, care, and cultural knowledge.

What interested me most while researching the cloth was that it is much more than just a material object. In many Pacific communities it has been used in ceremonies, funerals, gift exchanges, clothing, and celebrations. The patterns and markings on the cloth often carry cultural meaning and reflect traditions that have been passed down through generations.

The tapa samples in Alexander Shaw’s catalogue were collected during the voyages of Captain Cook in the eighteenth century and brought back to Britain. Because of this, the catalogue is connected not only to textile history, but also to colonial collecting and the way Pacific cultures were documented and represented in Europe.

One of the most interesting things about the Hunterian copy is that each version of the catalogue is slightly different. The tapa samples were inserted by hand, meaning every surviving copy is unique. This makes the object feel much more personal and material compared to an ordinary printed book. It is both a book and a textile archive at the same time, which is one of the reasons I was drawn to using it for this project.


### Why I Chose This Object

What drew me to this catalogue was how different it felt from a normal archival book. Even though it is presented as a printed publication, the inserted tapa samples make it feel much more physical and personal. Each fragment has visible textures, marks, fibres, and signs of age that cannot be fully captured through photographs alone. Looking through the catalogue made me think about how much of an object is lost when it becomes digital.

I was also interested in the fact that every surviving copy of the catalogue is slightly different because the cloth samples were inserted by hand. That idea of uniqueness felt important within a project about generative AI, especially since AI systems often work through repetition, datasets, and pattern recognition. 



### Research Question

This project asks whether generative AI can meaningfully interpret culturally significant archival textiles, or whether the process of digitisation reduces tapa cloth into visual patterns detached from their material and cultural context.

### Project Objectives
1. To explore how tapa cloth can be represented through digitisation and generative AI.
2. To compare AI-generated interpretations of tapa cloth with human and archival interpretations.
3. To investigate how texture, materiality, and cultural meaning are affected when archival objects become digital data.
4. To encourage audiences to question authenticity and representation within digital cultural heritage.
