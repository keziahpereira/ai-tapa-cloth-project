# Threads of Authenticity: Human–AI Interpretation of Alexander Shaw’s Tapa Cloth Catalogue



### AI for the Arts and Humanities (B)

This project explores the relationship between generative AI, digitisation, and cultural heritage through the study of *Sp Coll Hunterian K.5.22*, Alexander Shaw’s 1787 catalogue of tapa cloth samples collected during the voyages of Captain Cook.
The project investigates whether AI can meaningfully interpret culturally significant archival textiles, or whether digitisation reduces them to aesthetic surface patterns. It uses image analysis, AI-generated outputs, and human interpretation, the project examines questions of authenticity, colonial collecting, materiality, and archival representation.

# Introduction

### What is the object?

The selected object for this project is *Sp Coll Hunterian K.5.22 - A catalogue of the different specimens of cloth collected in the three voyages of Captain Cook, to the southern hemisphere* (London, 1787).

The catalogue contains samples of tapa cloth collected during Captain Cook’s voyages across the Pacific. Tapa cloth, also known as barkcloth, is a textile traditionally made in several Pacific cultures from the inner bark of trees. Unlike printed books, each surviving copy of the catalogue is materially unique because the textile samples were inserted by hand.

The object is held within the University of Glasgow Special Collections and represents an intersection between colonial collecting, material culture, archival preservation, and tactile history.
